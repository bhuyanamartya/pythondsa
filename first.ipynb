```python
name=input("Enter your name: ")
age=int(input("Enter your age: "))
if age<18:
  print(f"{name}, you're a child")
elif age>=18 and age<=35:
  print(f"{name}, you're a teenager")
elif age>=36 and age<=65:
  print(f"{name}, you're an adult")
else:
  print(f"{name}, you're old")
```

```python
famous_car_brands=list(('Mercedes','Toyota','Ferrari','Porsche','Lamborghini','Buggati'))
print(famous_car_brands)
famous_car_brands.insert(2,'Ford')
famous_car_brands.sort()
print(famous_car_brands)
print(famous_car_brands[0::2])
print(famous_car_brands[::-1])
print(len(famous_car_brands))
print(famous_car_brands.count('Ferrari'))
print(famous_car_brands.index('Porsche'))
```
